각 권역별 최다 사용 업종 명 및 해당 업종의 총 사용 금액.

```python
def top_use_info(data,region):
    region_df = data[data['시군구명'] == region]
    grouped_df = region_df.groupby('업종명칭')['총 거래금액'].sum()
    most_use_category = grouped_df.idxmax()
    max_amount = grouped_df.max()
    print(f'부산광역시 {region}의 최다 사용 업종명: {most_use_category}') 
    print(f"부산광역시 {region}의 '{most_use_category}'의 총 사용금액 {max_amount}"'\n\n')
print('1권역','\n')
top_use_info(dataDF1,'강서구')
top_use_info(dataDF1,'금정구')
top_use_info(dataDF1,'기장군')
top_use_info(dataDF1,'남구')
```
